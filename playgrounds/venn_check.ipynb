```python
# 例として2つの辞書を作成
dic1 = {'A': 1, 'B': 2, 'C': 3, 'D': 4}
dic2 = {'A': 4, 'B': 5, 'C': 6}

# 辞書を足し算でマージ
merged_dic = {key: dic1.get(key, 0) + dic2.get(key, 0) for key in dic1.keys() | dic2.keys()}

print(merged_dic)
```
