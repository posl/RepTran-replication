```python
import os, sys, time
import pickle
from tqdm.notebook import tqdm
import pandas as pd
import json
import numpy as np
from collections import defaultdict
from itertools import product
from scipy import stats
from datasets import load_from_disk
from transformers import ViTForImageClassification
from utils.helper import get_device, json2dict
from utils.vit_util import transforms, transforms_c100
from utils.constant import ViTExperiment
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("ticks")
import torch
device = get_device()
```

```python
def plot_tracker(key, tracker_n, alpha_list):
    """nを指定した時のalphaの違いによるtracker_n[key]の変化をプロットする"""
    colors = sns.color_palette("tab10", n_colors=len(alpha_list))
    plt.figure(figsize=(8, 6))
    for i, a in enumerate(alpha_list):
        color = colors[i]
        plt.plot(tracker_n[a][key], label=rf"$\alpha$={a}", color=color, alpha=0.7)
    plt.xlabel("generation")
    plt.ylabel(key)
    plt.title(f"{key} vs generation")
    plt.legend()
    plt.show()
    plt.close()
```

# 世代ごとのベストのfitnessの値を表示 (all)

```python
ds_name = "c100"
k = 0
tgt_rank = 1
misclf_type = "src_tgt"
# pretrained modelのロード
pretrained_dir = getattr(ViTExperiment, ds_name).OUTPUT_DIR.format(k=k)
de_dir = os.path.join(pretrained_dir, f"misclf_top{tgt_rank}", f"{misclf_type}_repair_weight_by_de")
if misclf_type == "all":
    de_dir = os.path.join(pretrained_dir, f"{misclf_type}_repair_weight_by_de")
tracker = defaultdict(dict)
n_list = [77]
alpha_list = [0.2, 0.4, 0.6, 0.8]
all_fitness_correct = []
all_fitness_incorrect = []
for n, alpha in product(n_list, alpha_list):
    tracker_path = os.path.join(de_dir, f"tracker_n{n}_alpha{alpha}.pkl")
    with open(tracker_path, "rb") as f:
        tracker[n][alpha] = pickle.load(f)
        met_keys = list(tracker[n][alpha].keys())
    all_fitness_correct.extend(tracker[n][alpha]["fitness_for_correct"])
    all_fitness_incorrect.extend(tracker[n][alpha]["fitness_for_wrong"])
# met_keysからall_pred_labelsを除外
met_keys.remove("all_pred_labels")
for n in n_list:
    tracker_n = tracker[n]
    for key in met_keys:
        plot_tracker(key, tracker_n, alpha_list)
```

```python
# all_fitness_correctを横軸, all_fitness_incorrectを縦軸にとった散布図
plt.figure(figsize=(8, 6))
plt.xlabel("fitness for correct")
plt.ylabel("fitness for incorrect")
plt.scatter(all_fitness_correct, all_fitness_incorrect, alpha=0.5)
plt.title("fitness for correct vs fitness for incorrect")
plt.show()
plt.close()
```

# FLのハイパラ $n$ と特定されるニューロンの割合
$n=768$ だと全ニューロン特定するので割合は 100% になる．

対象のニューロン数の母数は 768 \* 3072 + 768 \* 3072 = 4718592．
これに対して $n$ を指定することで， $8n^2$ のニューロンが特定されるので，百聞率での割合は，
$$
\frac{8n^2}{4718592} \times 100 (\%)
$$
となる．

```python
ratio = lambda n: 800 * n * n / 4718592
nlist, rlist = [], []
for n in range(5, 769):
    r = ratio(n)
    if r <= 100:
        nlist.append(n)
        rlist.append(r)
        if (n+5) % 10 == 0:
            print(f"n={n}, ratio={r:.3f}%")
plt.figure(figsize=(8, 6))
plt.plot(nlist, rlist)
plt.xlabel("n")
plt.ylabel("ratio (%)")
plt.title("n vs ratio")
plt.show()
```
